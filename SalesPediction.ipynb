### Simple Linear Regression Example

Let's create some dummy data and train a linear regression model to predict `y` from `x`.

```python
# Generate some sample data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Create a Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Visualize the results
plt.scatter(X, y, label='Original data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()

print(f"Intercept: {model.intercept_[0]:.2f}")
print(f"Coefficient: {model.coef_[0][0]:.2f}")
```

```python
df = pd.read_csv("sales.csv")

print(df.head())
```

```python
print(df.info())
```

```python
df["Day"] = range(len(df))
```

```python
import matplotlib.pyplot as plt

plt.plot(df["Day"], df["Sales"])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(df["Day"], df["Sales"])
plt.xlabel("Day")
plt.ylabel("Sales")
plt.title("Sales Trend")
plt.show()
```

```python
X = df[["Day"]]

y = df["Sales"]

print(X.head())
print(y.head())
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X, y)

print("Model trained successfully!")
```

```python
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
predicted_sales = model.predict([[10]])
print("Predicted Sales for Day 10:", predicted_sales[0])
```

```python
future_days = [[6], [7], [8], [9], [10]]

predictions = model.predict(future_days)

print(predictions)
```

```python
y_pred = model.predict(X)

for actual, predicted in zip(y, y_pred):
    print(f"Actual: {actual}, Predicted: {predicted:.2f}")
```

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X)

print("R² Score:", r2_score(y, y_pred))
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Day")
plt.ylabel("Sales")
plt.title("Linear Regression")
plt.show()
```
